```python
from collections import deque
# Read data from file
Readfile = open("data3.txt", "r")
Data = Readfile.readlines()
Readfile.close()

# Remove newline symbols from data
def DataCleaning(Data_lst):
    for i in range(len(Data_lst)):
        Data_lst[i] = Data_lst[i].strip('\n')
    return Data_lst

# Arrange data in list-of-list-of-dictionary
def DataRearrangement(Data_lst):
    newlist = list()
    templist = list()
    Dict = {}
    prev_Node = Data_lst[0].split(',', 1)[0]

    for Rows in Data_lst:
        Nodes = Rows.split(',', 1)
        line = Nodes[1].split('},', 1)
        line[0] = line[0] + '}'
        if prev_Node != Nodes[0]:
            newlist.append([prev_Node, Dict])
            Dict = {}
        Dict[line[0]] = line[1]
        prev_Node = Nodes[0]
    newlist.append([prev_Node, Dict])
    return newlist

Data = DataCleaning(Data)
Data = DataRearrangement(Data[:])

def getlist(node, data):
    rows = len(Data)
    cols = len(Data[0])
    for i in range(rows):
        if node == int(data[i][0]):
            return data[i][1]

def Cost(Ordering):
    cost = 0
    i = 0
    temp_lst = []
    cost_temp = []
    Ordering = Ordering[::-1]
    for order in Ordering:
        Plist = Ordering[i+1:]
        i += 1
        temp_lst = []
        cost_temp = []
        dic = getlist(order, Data)
        temp_lst.append(dic)
        for key, value in dic.items():
            val = key.split("{")
            val1 = val[1].split("}")
            val2 = val1[0].split(",")
            for j in range(len(val2)):
                if val2[j] == "":
                    cost_temp.append(float(value))
                    continue
                else:
                    val2[j] = int(val2[j])
            if set(val2).issubset(set(Plist)):
                cost_temp.append(float(value))
        cost_temp.sort()
        cost += cost_temp[0]
    print("Cost:", round(cost, 2))
    return round(cost, 2)

Order_lst = [1, 2, 3, 4, 5]
Cost(Order_lst)

# Define the graph based on the data
graph = {}
for entry in Data:
    node = int(entry[0])
    adjacent_nodes = []
    for key in entry[1]:
        adjacent_nodes.extend([int(x) for x in key.strip('{}').split(',') if x.strip()])
    graph[node] = adjacent_nodes

# Breadth-First Search (BFS) algorithm
def bfs(graph, start):
    visited = set()
    queue = deque([start])
    traversal_order = []

    while queue:
        node = queue.popleft()
        if node not in visited:
            traversal_order.append(node)
            visited.add(node)
            queue.extend(graph[node])

    return traversal_order

# Depth-First Search (DFS) algorithm
def dfs(graph, start):
    visited = set()
    stack = [start]
    traversal_order = []

    while stack:
        node = stack.pop()
        if node not in visited:
            traversal_order.append(node)
            visited.add(node)
            stack.extend(reversed(graph[node]))

    return traversal_order

# Example usage of BFS
start_node = 1  # Start BFS from node 1
bfs_traversal_order = bfs(graph, start_node)
print("BFS Traversal Order:", bfs_traversal_order)

# Example usage of DFS
dfs_traversal_order = dfs(graph, start_node)
print("DFS Traversal Order:", dfs_traversal_order)
```
